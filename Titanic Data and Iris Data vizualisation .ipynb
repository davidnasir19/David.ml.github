```python
titanic = sns.load_dataset("titanic")
```

```python
titanic.head()
```

```python
titanic.drop(["embarked","class","who","adult_male","alive","alone"], axis = 1,inplace = True)
titanic.dropna(axis = 0, inplace = True)
titanic.head()
```

```python
sns.pairplot(titanic)
```

```python
sns.catplot(x = "pclass", y = "age", hue = "sex", data = titanic)
```

```python
    sns.boxplot(x = "pclass", y = "age", hue = "sex", data = titanic)
```

```python
sns.distplot(titanic["fare"])
plt.show()
```

```python
sns.jointplot(x="age",y="fare", data = titanic, kind = "hex")
```

```python
sns.jointplot(x="age",y="fare", data = titanic, kind = "kde", fill = True)
```

```python
sns.heatmap(titanic.corr(numeric_only=True))
```
