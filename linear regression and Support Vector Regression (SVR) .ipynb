```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
m = 100
x = np.linspace(0,10,m).reshape(m,1)
y = x + np.random.randn(m,1)
```

```python
plt.scatter(x,y)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(x,y)
model.score(x,y)
```

```python
prediction = model.predict(x)
```

```python
plt.scatter(,y)
plt.plot(x, prediction, c = "r")
```

```python
c = 100
a = np.linspace(0,10,c).reshape(c,1)
b = a**2 + np.random.randn(c,1)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
```

```python
model = SVR(C=100)
model.fit(a,b)
model.score(a,b)
```

```python
prediction = model.predict(a)
```

```python
plt.scatter(a,b)
plt.plot(a, prediction, c = "r", lw = 3)
```
